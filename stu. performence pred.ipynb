```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
for col in df.select_dtypes(include=[object]):
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
df['present'] = 94 - df['absences']
```

```python
features = df.iloc[:,:-3]
target = df.iloc[:,-3:]
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
X_train,X_test, y_train,y_test = train_test_split(features,target, test_size=0.15,random_state=33)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test= scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.pipeline import Pipeline
```

```python
models = [LinearRegression(),RandomForestRegressor()]

for model in models:
    model.fit(X_train,y_train)
    
    prediction = model.predict(X_test)
    RMSE = mean_squared_error(y_test,prediction)
    MAE = mean_absolute_error(y_test,prediction)

    print(f'{model} : ')
    print('Training Accuracy : ', metrics.r2_score(y_train, model.predict(X_train)))
    print('Validation Accuracy : ', metrics.r2_score(y_test, model.predict(X_test)))
    print('RMSE :',RMSE)
    print(f'MAE :',MAE)
    print()
```

```python
parameter_grid ={
    'lr__fit_intercept': [True, False], 
    'lr__positive': [True, False]       
}

pipeline_lr = Pipeline([
    ('scaler',StandardScaler()),
    ('lr',LinearRegression())
])

clf = GridSearchCV(estimator=pipeline_lr,param_grid=parameter_grid, cv=5,verbose=True,n_jobs=-1)
clf
```

```python
clf = clf.fit(X_train,y_train)
print("Best Parameters for LinearRegression:", clf.best_params_)
print("Best R2 Score for LinearRegression:", clf.best_score_)
```

```python
print(f'Accuracy: {clf.score(X_train,y_train)}')
```
